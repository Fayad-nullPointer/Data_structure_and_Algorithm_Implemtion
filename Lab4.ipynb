### Merge Sort

```python
def merge_sort(arr : list ,start,end):
    if end > start :
        mid = (start+end)//2
        merge_sort(arr , start,mid)
        merge_sort(arr,mid+1 ,end)
        merge(arr, start,mid,end)
def merge(arr : list ,start ,mid ,end):
    list1 = start
    list2 = mid+1
    temp = []
    while list1 <= mid and list2 <= end:
        if arr [list1]< arr[list2]:
            temp.append(arr[list1])
            list1 +=1
        else:
            temp.append(arr[list2])
            list2+=1
    # Move remonder from both lists to temp
    while list1 <mid:
        temp.append(arr[list1])
        list1 +=1
    while list2 <= end:
        temp.append(arr[list2])
        list2+=1
    arr = temp
    del temp
    # Copy the sorted elements from temp back to the original array
    for i in range(len(temp)):
        arr[start + i] = temp[i]
```

```python
def min_max(arr,start,end):
    # 1 element
    mid = (start+end)//2
    if start == end:
        return arr
    # 2 element 
    elif start == end-1:
        if arr[start] < arr[end]:
            min = arr[start]
            max = arr[end]
            return min ,max
        else :
            min = arr[end]
            max = arr[start]
            return min ,max
    else :
        mid = (start + end) // 2
        min_left,max_left = min_max(arr,start,mid)
        min_right,max_right = min_max(arr,mid+1,end)
        final_min = min(min_left,min_right)
        final_max = max(max_left,max_right)
        return final_min,final_max
```

```python
class knapsack_package:
    def __init__(self, w:int,val :int):
        self.weight = w
        self.value = val
        self.cost = self.weight/self.value
    def get_weight(self):
        return self.weight
    def get_value(self):
        return self.value
    def get_cost(self):
        return self.cost
```

```python
def knapsack_process(weights : list , values :list ,m :int,n :int):
    knapscks = []
    for i in range(n):
        knapscks.append(knapsack_package(weights[i],values[i]))
    knapscks = sorted(knapscks , key= lambda x:x.get_cost() ,reverse= True)
    remain = m
    result = 0
    i = 0
    stopProc = False
    while(not stopProc):
        if knapscks[i].get_weight() < remain:
            remain -=knapscks[i].get_weight()
            result += knapscks[i].get_value()
            print(f"Pack {i} Weight {knapscks[i].get_weight()} Value {knapscks[i].get_value()}")
            i+=1
        elif remain >0:
            res = (knapscks[i].get_value()*remain)/knapscks[i].get_weight()
            result += res
            remain = 0
            print(f"Pack {i} Weight {knapscks[i].get_weight()} Value {knapscks[i].get_value()}")
            stopProc =True
        else :
            stopProc =True
    print(f"max value = {result}")    
```

```python
#try knapsack_process([10,20,30],[60,100,120],50,3)
weights = [10, 20, 30]
values = [60, 100, 120]
capacity = 50
n = len(weights)
knapsack_process(weights, values, capacity, n)
```
