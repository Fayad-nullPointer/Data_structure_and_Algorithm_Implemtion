```python
class student:
    """
    A class to represent a student.

    Attributes
    ----------
    id : int
        The unique identifier for the student.
    name : str
        The name of the student.
    grade : list[float]
        A list of grades the student has achieved.
    """

    def __init__(self, id: int, name: str, grade: list[float]):
        """
        Constructs all the necessary attributes for the student object.

        Parameters
        ----------
        id : int
            The unique identifier for the student.
        name : str
            The name of the student.
        grade : list[float]
            A list of grades the student has achieved.
        """
        self.id = id
        self.name = name
        self.grade = grade

    # Implement Getters and Setters
    def get_id(self) -> int:
        """
        Gets the student's ID.

        Returns
        -------
        int
            The ID of the student.
        """
        return self.id

    def set_id(self, id: int):
        """
        Sets the student's ID.

        Parameters
        ----------
        id : int
            The new ID for the student.
        """
        self.id = id

    def get_name(self) -> str:
        """
        Gets the student's name.

        Returns
        -------
        str
            The name of the student.
        """
        return self.name

    def set_name(self, name: str):
        """
        Sets the student's name.

        Parameters
        ----------
        name : str
            The new name for the student.
        """
        self.name = name

    def get_grade(self) -> list[float]:
        """
        Gets the student's grades.

        Returns
        -------
        list[float]
            The list of grades the student has achieved.
        """
        return self.grade

    def set_grade(self, grade: list[float]):
        """
        Sets the student's grades.

        Parameters
        ----------
        grade : list[float]
            The new list of grades for the student.
        """
        self.grade = grade

    def get_average(self) -> float:
        """
        Calculates and returns the average of the student's grades.

        Returns
        -------
        float
            The average of the grades. Returns an error if the grade list is empty.
        """
        return sum(self.grade) / len(self.grade)
```

```python
class Node:
    def __init__(self , data : student):
        self.data = data
        self.prev = None
        self.next = None
```

```python
class DLL:
    """
    A class to represent a Doubly Linked List for student objects.
    """
    
    def __init__(self):
        """Initializes an empty Doubly Linked List with null head and tail."""
        self.head = None
        self.tail = None

    def append(self, stu: student):
        """
        Adds a new student to the end (tail) of the list.
        """
        nd = Node(stu)
        if self.head is None:
            # If the list is empty, both head and tail must point to the new node.
            self.head = nd
            self.tail = nd
            print("Adding Done")
        else:
            # Append means adding to the tail, not the head.
            nd.prev = self.tail
            self.tail.next = nd
            self.tail = nd
            print("Adding Done")

    def prepend(self, stu: student):
        """
        Adds a new student to the beginning (head) of the list.
        """
        nd = Node(stu)
        if self.head is None:
            # Handle empty list cleanly.
            self.head = nd
            self.tail = nd
            print("Adding Done")
        else:
            # Prepend means adding to the head, not the tail.
            nd.next = self.head
            self.head.prev = nd
            self.head = nd
            print("Adding Done")

    def delete_by_id(self, id: int):
        """
        Removes a student node from the list using their ID.
        """
        temp = self.head  
        # as long as temp exists and the ID does not match.
        while temp and temp.data.get_id() != id:
            temp = temp.next
            
        if not temp:
            print("Not Found")
        else:
            # Handle all cases (node is head, node is tail, node is in the middle)
            if temp.prev:
                temp.prev.next = temp.next
                
            else:
                self.head = temp.next  # We are deleting the head

            if temp.next:
                temp.next.prev = temp.prev
            else:
                self.tail = temp.prev  # We are deleting the tail
                
            print("Deleted Done")

    def search_by_name(self, name: str):
        """
        Searches for a student by their name.
        Returns the Node if found, or None if not found.
        """
        temp = self.head
        # Fix: Prevent AttributeError if list is empty or reaches the end without finding.
        # Loop until we run out of nodes or find the requested name.
        while temp and temp.data.get_name() != name:
            temp = temp.next
            
        return temp # Returns the matching node, or None if it simply reached the end.

    def search_by_ID(self, id: int):
        """
        Searches for a student by their ID.
        Returns the Node if found, or None if not found.
        """
        temp = self.head
        # Fix: Prevent AttributeError. Traverse until the ID is found or end of list.
        while temp and temp.data.get_id() != id:
            temp = temp.next
            
        return temp

    def count_nodes(self) -> int:
        """
        Counts and returns the total number of students in the list.
        """
        temp = self.head
        count = 0
        while temp:
            count += 1
            temp = temp.next
        return count

    def display_forword(self):
        """
        Prints the details of all students in the list from head to tail.
        """
        temp = self.head
        if not temp:
            print("Empty list")
        else:
            while temp:
                print(f"Name {temp.data.get_name()} ID {temp.data.get_id()}")
                # Fix: Added traversal step to prevent an infinite loop.
                temp = temp.next 

    def display_backword(self):
        """
        Prints the details of all students in the list in reverse order (from tail to head).
        """
        temp = self.tail
        if not temp:
            print("Empty list")
        else:
            while temp:
                print(f"Name {temp.data.get_name()} ID {temp.data.get_id()}")
                # Fix: Added traversal step backward (using `.prev`) to prevent an infinite loop.
                temp = temp.prev
```
