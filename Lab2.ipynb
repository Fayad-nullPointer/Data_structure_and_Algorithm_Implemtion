```python
class student:
    """
    A class to represent a student.

    Attributes
    ----------
    id : int
        The unique identifier for the student.
    name : str
        The name of the student.
    grade : list[float]
        A list of grades the student has achieved.
    """

    def __init__(self, id: int, name: str, grade: list[float]):
        """
        Constructs all the necessary attributes for the student object.

        Parameters
        ----------
        id : int
            The unique identifier for the student.
        name : str
            The name of the student.
        grade : list[float]
            A list of grades the student has achieved.
        """
        self.id = id
        self.name = name
        self.grade = grade

    # Implement Getters and Setters
    def get_id(self) -> int:
        """
        Gets the student's ID.

        Returns
        -------
        int
            The ID of the student.
        """
        return self.id

    def set_id(self, id: int):
        """
        Sets the student's ID.

        Parameters
        ----------
        id : int
            The new ID for the student.
        """
        self.id = id

    def get_name(self) -> str:
        """
        Gets the student's name.

        Returns
        -------
        str
            The name of the student.
        """
        return self.name

    def set_name(self, name: str):
        """
        Sets the student's name.

        Parameters
        ----------
        name : str
            The new name for the student.
        """
        self.name = name

    def get_grade(self) -> list[float]:
        """
        Gets the student's grades.

        Returns
        -------
        list[float]
            The list of grades the student has achieved.
        """
        return self.grade

    def set_grade(self, grade: list[float]):
        """
        Sets the student's grades.

        Parameters
        ----------
        grade : list[float]
            The new list of grades for the student.
        """
        self.grade = grade

    def get_average(self) -> float:
        """
        Calculates and returns the average of the student's grades.

        Returns
        -------
        float
            The average of the grades. Returns an error if the grade list is empty.
        """
        return sum(self.grade) / len(self.grade)
```

```python
class Node:
    def __init__(self,val):
        self.value = val
        self.prev = None
        self.next = None
```

```python
class StackLL:
    """
    A class to represent a Stack using a Linked List.
    """
    
    def __init__(self):
        """
        Initializes an empty Stack.
        """
        self.head = None

    def push(self, stu: student):
        """
        Pushes a new student onto the top of the stack.

        Parameters
        ----------
        stu : student
            The student object to be added to the stack.
        """
        nd = Node(stu)
        # Point the new node's next to the current head
        nd.next = self.head
        # Make the new node the head of the stack
        self.head = nd

    def pop(self):
        """
        Removes and returns the top node from the stack.

        Returns
        -------
        Node or None
            The top node of the stack, or None if the stack is empty.
        """
        if self.is_empty():
            return None
            
        nd = self.head
        self.head = self.head.next
        
        # Disconnect the popped node from the stack
        nd.next = None
        return nd

    def peek(self):
        """
        Returns the top node of the stack without removing it.

        Returns
        -------
        Node or None
            The top node of the stack, or None if the stack is empty.
        """
        return self.head

    def is_empty(self) -> bool:
        """
        Checks if the stack is empty.

        Returns
        -------
        bool
            True if the stack is empty, False otherwise.
        """
        return self.head is None

    def size(self) -> int:
        """
        Calculates the number of elements in the stack.

        Returns
        -------
        int
            The number of elements in the stack.
        """
        counter = 0
        current = self.head
        
        # Iterate using a temporary pointer to avoid modifying self.head
        while current:
            counter += 1
            current = current.next
            
        return counter
```

```python
class Queue:
    """
    A class to represent a Queue using a Python list with a fixed size.
    """
    
    def __init__(self, size: int):
        """
        Initializes an empty Queue with a specific maximum size.
        
        Parameters
        ----------
        size : int
            The maximum number of elements the queue can hold.
        """
        self.queue = []
        self.size = size

    def enqueue(self, stu: student):
        """
        Adds a student to the rear of the queue.
        
        Parameters
        ----------
        stu : student
            The student object to be added to the queue.
        """
        if self.is_full():
            print("Queue is full!")
            return
            
        # A list-based queue does not need Node wrappers like a Linked List.
        # We can store the student object directly.
        self.queue.append(stu)

    def dequeue(self):
        """
        Removes and returns the element at the front of the queue.
        
        Returns
        -------
        student or None
            The student object at the front, or None if the queue is empty.
        """
        if self.is_empty():
            return None
            
        return self.queue.pop(0)

    def front(self):
        """
        Returns the element at the front of the queue without removing it.
        
        Returns
        -------
        student or None
            The student at the front, or None if the queue is empty.
        """
        if self.is_empty():
            return None
            
        return self.queue[0]

    def is_empty(self) -> bool:
        """
        Checks if the queue is empty.
        
        Returns
        -------
        bool
            True if the queue is empty, False otherwise.
        """
        return len(self.queue) == 0

    def is_full(self) -> bool:
        """
        Checks if the queue has reached its maximum size.
        
        Returns
        -------
        bool
            True if the queue is full, False otherwise.
        """
        return len(self.queue) == self.size

    def size(self) -> int:
        """
        Returns the current number of elements in the queue.
        
        Returns
        -------
        int
            The number of elements in the queue.
        """
        return len(self.queue)
```
