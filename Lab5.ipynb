```python
class Node:
    def __init__(self ,val):
        self.value = val
        self.left = None
        self.right = None
```

```python
class Binary_tree:
    def __init__(self):
        self.root = None
    def insert_helper(self,root : Node ,val :int):
        nd = Node(nd)
        if not root:
            root = nd
        else:
            if val < root.value:
                if root.left == None:
                    root.left = nd
                else:
                    self.insert_helper(root.left,val)
            elif val > root.value:
                if root.right ==None:
                    root.right = nd
                else:
                    self.insert_helper(root.right,val)
    def insert(self,val):
        nd =Node(val)
        if self.root ==None:
            self.root = val
        else:
            self.insert_helper(self.root,val)
```

```python
def Inorder_traversal(root : Node):
    if root == None:
        return
    Inorder_traversal(root.left)
    print(root.value)
    Inorder_traversal(root.right)
def preorder_traversal(root : Node):
    if root == None:
        return
    print(root.value)
    preorder_traversal(root.left)
    preorder_traversal(root.right)
def postorder_traversal(root : Node):
    if root == None:
        return
    print(root.value)
    postorder_traversal(root.left)
    postorder_traversal(root.right)
```

```python
def InOrder_PreOrder(In: list, pre: list):
    # Base case
    if not In or not pre:
        return None

    # Root
    root_value = pre[0]
    root_node = Node(root_value)

    # Find root in inorder
    index_of_root = In.index(root_value)

    # Split inorder
    left_inorder = In[:index_of_root]
    right_inorder = In[index_of_root + 1:]

    # Split preorder (skip root)
    left_preorder = pre[1: 1 + len(left_inorder)]
    right_preorder = pre[1 + len(left_inorder):]

    # Recursion
    root_node.left = InOrder_PreOrder(left_inorder, left_preorder)
    root_node.right = InOrder_PreOrder(right_inorder, right_preorder)

    return root_node
```

```python
def InOrder_PostOrder(In: list, post: list):
    # Base case
    if not In or not post:
        return None

    # Root
    root_value = post[-1]
    root_node = Node(root_value)

    # Find root in inorder
    index_of_root = In.index(root_value)

    # Split inorder
    left_inorder = In[:index_of_root]
    right_inorder = In[index_of_root + 1:]

    # Split postorder (exclude last element = root)
    left_postorder = post[:len(left_inorder)]
    right_postorder = post[len(left_inorder):-1]

    # Recursion
    root_node.left = InOrder_PostOrder(left_inorder, left_postorder)
    root_node.right = InOrder_PostOrder(right_inorder, right_postorder)

    return root_node
```

```python
class Graph:
    def __init__(self):
        self.graph = {}

    def add_vertex(self, v):
        if v not in self.graph:
            self.graph[v] = []

    def add_edge(self, u, v, undirected=False):
        # Ensure both vertices exist
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []

        # Add edge u → v
        if v not in self.graph[u]:
            self.graph[u].append(v)

        # If undirected, also add v → u
        if undirected:
            if u not in self.graph[v]:
                self.graph[v].append(u)
```

```python
def DFS_Recurive(start_node, Graph, visited=None, result=None):
    # Initialize visited and result only on the very first call
    if visited is None:
        visited = set()
    if result is None:
        result = []
        
    graph_dict = Graph.graph
    
    # Process the node
    if start_node not in visited:
        result.append(start_node)
        visited.add(start_node)
    else:
        return result
    
    # Recursively visit neighbors
    if start_node in graph_dict:
        for neighbour in graph_dict[start_node]:
            DFS_Recurive(neighbour, Graph, visited, result)
            
    return result
```

```python
def Iterative_DFS(start_node ,Graph : Graph):
    visted = set()
    result = []
    stack = [start_node]
    graph_dict = Graph.graph
    while stack:
        current = stack.pop()

        visted.add(current)
        result.append(current)

        if current in graph_dict:
            for neighoubr in graph_dict[current]:
                if neighoubr not in visted:
                    stack.append(neighoubr)
    return result
```

```python
def BFS(start_node , Graph :Graph):
    visited = set()
    result = []
    Queue = [start_node]
    graph_dict  = Graph.graph

    while Queue:
        current = Queue.pop(0)
        if current not in visited:
            visited.add(current)
            Queue.append(current)

        if current in graph_dict:
            for neighoubr in graph_dict[current]:
                if neighoubr not in visited:
                    Queue.append(neighoubr)
    return result
```
